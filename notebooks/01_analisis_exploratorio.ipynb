## Análisis de Ventas por Categoría

Este análisis tiene como objetivo identificar las categorías de productos 
con mayor volumen de ventas, con el fin de detectar oportunidades 
de negocio y enfocar estrategias comerciales.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import os

USER = 'root'
PASSWORD = os.getenv('DB_PASSWORD')
HOST = 'localhost'
PORT = '3306'
DATABASE = 'analisis_ventas'    

engine = create_engine(f'mysql+pymysql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}')
```

## Consulta SQL

Se agrupan las ventas por categoría de producto y se calcula el total de ventas por cada una.

```python
query = """
SELECT Product_Category, SUM(Sales_Amount) AS Total_Ventas
FROM ventas
GROUP BY Product_Category
ORDER BY Total_Ventas DESC;
"""
```

## Visualización

Se representa un gráfico de barras para comparar las categorías según su nivel de ventas.

```python
df = pd.read_sql(query, engine)

plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Product_Category', y='Total_Ventas')
plt.title('Ventas por Categoria de Producto')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Insight

Se observa que ciertas categorías concentran la mayor parte de las ventas, lo que indica una oportunidad para potenciar estos productos o replicar estrategias en otras categorías.
